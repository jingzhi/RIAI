# Zonotope Abstract Domain

[ref1](http://math.sfsu.edu/beck/papers/zonotopes.pdf)

__polytopes__ : a geometric object with flat sides in nD. 
* 3d -> polyhedron
* 2d -> polygon
* 1d -> line segment
* kd -> k-polytope

Flat sides mean that the sides of a (k+1)-polytope consist of k-polytopes that may have (k-1)-polytopes in common. 

Considering d-polytopes $P_1,P_2,...P_n \in \mathbb{R}^d$.

Each of them can have arbitary number of line segments/vertices. Eache vertex will be a d-dimension vector.

In 2D, triangle described by threevertices. rectangle described by two vertices $[a,b] \times [c,d]$

The __Minkowski Sum__  of two sets A,B in a vector space is defined as: 
$$A+B := \{a+b: a \in A,b \in B\}$$

The __Minkowski Sum__ of n d-polytopes are defined as:
$$(P_1+P_2+...+P_n) :=\{\mathbf{x}_1+\mathbf{x}_2+...+ \mathbf{x}_n :\mathbf{x}_j \in P_j\}$$
The result of the addition is formed by adding the subset  in each polytope.

Note zero vector is an identity element: $S+\{0\}=S$.

The empty set annihilates every other subset: $S+\emptyset = \emptyset$

__Newton polytope__ : Consider polynomial with d-variables $p(z_1,z_2,...z_d)$. The Newton polytope of this polynomial is sthe [convex hull](https://en.wikipedia.org/wiki/Convex_hull) of p (the smallest convex set that contains p). 

$$N(pq)=N(p)+N(q),$$where the sum is the Minkowski sum.

A __zonotope__ P in $ \mathbb{R}^d $ is a convex polytope whch can be represented as Minkowski sum of finite line segments:
$$ P= L_1+L_2+...+L_m $$

$L_i$ is the line segment $[a_i,b_i]$ defined by two points $a_i,b_i \in \mathbb{R}^d$ as end points, all points on the line is considered to be in $L_i$. These m line segments are called the __generators__ of the zonotope P.

[Examples](https://www.cs.mcgill.ca/~fukuda/760B/handouts/expoly3.pdf)


Now we consider the line segments with one end tied to origin.

The __Minkowski Sum__ of n line segments in $\mathbb{R}^d$, which has one end fixed to origin, the other located at vector $\mathbf{u}_i \in \mathbb{R}^d$ is defined as:
$$Z(\mathbf{u}_1,\mathbf{u}_2,..., \mathbf{u}_n) :=\{\mathbf{x}_1+\mathbf{x}_2+...+ \mathbf{x}_n :\mathbf{x}_j =\gamma_j \mathbf{u}_j \text{ with } \gamma_j \in [0,1] \}$$

$$\begin{align}
Z(\mathbf{u}_1,\mathbf{u}_2,..., \mathbf{u}_n) &= \{ \gamma_1 \mathbf{u}_1 + \gamma_2\mathbf{u}_2 +...+ \gamma_n\mathbf{u}_n : 0 \leq \gamma_j \leq 1 \}\\
&=\{(\mathbf{u}_1 \mathbf{2}_2 ... \mathbf{u}_n)[0,1]^n\}\\
&=\mathbf{A}[0,1]^n
\end{align}$$


Note matrix __A___ has all line segment vectors as its column vectors. A __zonotope__ is defined to be any translate of $\mathbf{A}[0,1]^n$.
That is 
$$\mathbf{A}[0,1]^n+\mathbf{b}$$
for some point $\mathbf{b} \in \mathbb{R}^d$.

It has a center of mass which can be translated to orgin (with a larger image) by:
$$\begin{align}
&2\mathbf{A}[0,1]^n-(\mathbf{u}_1,..., \mathbf{u}_n)\\
=&\{ 2\gamma_1 \mathbf{u}_1 + ...+ 2\gamma_n\mathbf{u}_n -(\mathbf{u}_1,..., \mathbf{u}_n): 0 \leq \gamma_j \leq 1 \}\\
=&\{ (2\gamma_1-1) \mathbf{u}_1 + ...+ (2\gamma_n-1)\mathbf{u}_n: 0 \leq \gamma_j \leq 1 \}\\
=&\mathbf{A}[-1,1]^n
\end{align}$$
